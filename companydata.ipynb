```python
from apyori import apriori
records = []
for i in range(0, 420):
    records.append([str(data.values[i,j]) for j in range(1, 10)])
    print(records)
```

```python
    


association_rules = apriori(records, min_support=0.3, min_confidence=0.7,min_lift=1.2,min_len=2)

association_results = list(association_rules)
for i in association_results:
    print(' ')
    print(i)
```

```python
result=[list(association_results[i][0]) for i in range(0,len(association_results))]
result
```
